```python
import giancarlo as gc

Qu = gc.Symbol('Q_u')
Qd = gc.Symbol('Q_d')
e = gc.Symbol('e')

u, ubar = gc.QuarkField('u')
d, dbar = gc.QuarkField('d')
s, sbar = gc.QuarkField('s')
c, cbar = gc.QuarkField('c')
```

```python
def O1(x):
    sa, sb, sc, sd = gc.default.var(), gc.default.var(), gc.default.var(), gc.default.var()
    ca, cb, cc, cd = gc.default.var(), gc.default.var(), gc.default.var(), gc.default.var()
    return ubar(x, sa, ca) * gc.DiracGamma('L', sa, sb) * d(x, sb, cb) * sbar(x, sc, cc) * gc.DiracGamma('L', sc, sd) * c(x, sd, cd)

def O1dag(x):
    sa, sb, sc, sd = gc.default.var(), gc.default.var(), gc.default.var(), gc.default.var()
    ca, cb, cc, cd = gc.default.var(), gc.default.var(), gc.default.var(), gc.default.var()
    return dbar(x, sa, ca) * gc.DiracGamma('L', sa, sb) * u(x, sb, cb) * cbar(x, sc, cc) * gc.DiracGamma('L', sc, sd) * s(x, sd, cd)

O1('x')
```

```python
corr = O1('x') * O1dag('y')
corr.wick()
```
