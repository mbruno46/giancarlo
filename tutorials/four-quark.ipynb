```python
import giancarlo as gc

Qu = gc.Symbol('Q_u')
Qd = gc.Symbol('Q_d')
e = gc.Symbol('e')

u, ubar = gc.QuarkField('u')
d, dbar = gc.QuarkField('d')
s, sbar = gc.QuarkField('s')
c, cbar = gc.QuarkField('c')

def Jgamma(x, mu):
    sa, sb = gc.default.var(), gc.default.var()
    c = gc.default.var()
    up = Qu * ubar(x, sa, c) * gc.DiracGamma(mu, sa, sb) * u(x, sb, c)
    down = Qd * dbar(x, sa, c) * gc.DiracGamma(mu, sa, sb) * d(x, sb, c)
    return up + down

Jgamma('x', r'\mu')
```

```python
hvp = Jgamma('x', r'\mu') * Jgamma('y', r'\nu')
hvp.wick().trace(['spin','color']).simplify()
```

```python
def O1(x, mu):
    sa, sb, sc, sd = gc.default.var(), gc.default.var(), gc.default.var(), gc.default.var()
    ca, cb, cc, cd = gc.default.var(), gc.default.var(), gc.default.var(), gc.default.var()
    return ubar(x, sa, ca) * d(x, sb, cb) * sbar(x, sc, cc) * c(x, sd, cd)

def O1dag(x, mu):
    sa, sb, sc, sd = gc.default.var(), gc.default.var(), gc.default.var(), gc.default.var()
    ca, cb, cc, cd = gc.default.var(), gc.default.var(), gc.default.var(), gc.default.var()
    return dbar(x, sa, ca) * u(x, sb, cb) * cbar(x, sc, cc) * s(x, sd, cd)

O1('x', r'\mu')
```

```python
corr = O1('x', r'\mu') * O1dag('y', r'\nu')
corr.wick()
```
